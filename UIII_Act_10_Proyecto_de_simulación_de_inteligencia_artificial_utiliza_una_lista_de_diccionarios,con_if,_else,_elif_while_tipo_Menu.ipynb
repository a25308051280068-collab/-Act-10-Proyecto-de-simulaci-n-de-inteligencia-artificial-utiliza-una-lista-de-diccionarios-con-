```python
# --- CONFIGURACIÓN INICIAL ---
# Empezamos con la lista de psicólogos vacía y el diccionario del paciente vacío
los_psicologos = []
paciente_actual = {}

print("=========================================")
print("🤖 BIENVENIDO AL SIMULADOR DE PSICÓLOGOS IA 🤖")
print("=========================================")

# --- CICLO PRINCIPAL (MENÚ INTERACTIVO) ---
while True:
    print("\n--- MENÚ PRINCIPAL ---")
    print("1. Registrar nuevo paciente")
    print("2. Ver pacientes registrados")
    print("3. Iniciar sesión de terapia con IA")
    print("4. Salir del programa")

    opcion = input("\nSelecciona una opción (1-4): ")

    # OPCIÓN 1: REGISTRAR PACIENTE
    if opcion == "1":
        print("\n--- REGISTRO DE PACIENTE ---")
        nombre = input("Nombre del paciente: ")
        edad = input("Edad del paciente: ")
        emocion = input("¿Cómo se siente hoy? (triste / enojado / feliz): ")

        # Llenamos el diccionario vacío con los datos del paciente
        paciente_actual = {
            "nombre": nombre,
            "edad": edad,
            "emocion_inicial": emocion
        }

        # Agregamos el diccionario a la lista de psicólogos (historial)
        los_psicologos.append(paciente_actual)
        print(f"✅ ¡Paciente {nombre} registrado con éxito en el sistema!")

    # OPCIÓN 2: VER PACIENTES
    elif opcion == "2":
        print("\n--- HISTORIAL DE PACIENTES EN LA CLÍNICA ---")

        if los_psicologos == []:
            print("❌ No hay pacientes registrados todavía.")
        else:
            # Usamos un ciclo for para recorrer la lista de diccionarios
            contador = 1
            for paciente in los_psicologos:
                print(f"\nPaciente N° {contador}:")
                print(f"- Nombre: {paciente['nombre']}")
                print(f"- Edad: {paciente['edad']} años")
                print(f"- Emoción reportada: {paciente['emocion_inicial']}")
                contador = contador + 1

    # OPCIÓN 3: SIMULADOR DE IA (TERAPIA)
    elif opcion == "3":
        print("\n--- CHATBOT PSICÓLOGO IA ---")

        if los_psicologos == []:
            print("❌ Primero debes registrar al menos un paciente (Opción 1).")
        else:
            nombre_chat = input("Introduce el nombre del paciente que hablará con la IA: ")

            # Buscamos si el paciente existe en nuestra lista de diccionarios
            encontrado = False
            for paciente in los_psicologos:
                if paciente["nombre"] == nombre_chat:
                    encontrado = True

            if encontrado == True:
                print(f"\n🤖 Hola {nombre_chat}, soy tu psicólogo virtual.")
                print("Cuéntame cómo te ha ido hoy (Escribe 'salir' para terminar la terapia):")
```
