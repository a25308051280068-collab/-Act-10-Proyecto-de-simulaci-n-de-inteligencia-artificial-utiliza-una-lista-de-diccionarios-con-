```python
# --- CONFIGURACIÓN INICIAL ---
# Las dos únicas variables solicitadas empiezan vacías
psicologos = {}
lospsicologo = []

print("=========================================")
print("🤖 BIENVENIDO AL SISTEMA DE PSICÓLOGOS IA 🤖")
print("=========================================")

# --- CICLO PRINCIPAL (MENÚ INTERACTIVO) ---
while True:
    print("\n--- MENÚ PRINCIPAL ---")
    print("1. Registrar nuevo psicólogo")
    print("2. Ver lista de psicólogos")
    print("3. Activar simulación de respuesta IA")
    print("4. Salir del programa")

    opcion = input("\nSelecciona una opción (1-4): ")

    # OPCIÓN 1: REGISTRAR PSICÓLOGO
    if opcion == "1":
        print("\n--- REGISTRO DE PSICÓLOGO ---")
        nombre = input("Nombre del psicólogo: ")
        especialidad = input("Especialidad (clínica / educativa / laboral): ")
        estado = input("¿Cómo se siente hoy? (triste / enojado / feliz): ")

        # Llenamos el diccionario 'psicologo' con sus llaves y valores
        psicologo = {
            "nombre": nombre,
            "especialidad": especialidad,
            "emocion_actual": estado
        }

        # Agregamos el diccionario 'psicologo' a la lista 'lospsicologos'
        lospsicologos.append(psicologo)
        print(f"✅ ¡Psicólogo {nombre} registrado con éxito!")

    # OPCIÓN 2: VER LOSPSICOLOGOS
    elif opcion == "2":
        print("\n--- LISTA DE PSICÓLOGOS REGISTRADOS (lospsicologos) ---")
        if lospsicologos == []:
            print("❌ La lista 'lospsicologos' está vacía todavía.")
        else:
            contador = 1
            # El ciclo FOR recorre directamente cada 'psicologo' en la lista 'lospsicologos'
            for psicologo in lospsicologos:
                print(f"\nPsicólogo N° {contador}:")
                print(f"- Nombre: {psicologo['nombre']}")
                print(f"- Especialidad: {psicologo['especialidad']}")
                print(f"- Estado de ánimo: {psicologo['emocion_actual']}")
                contador = contador + 1

    # OPCIÓN 3: SIMULADOR DE INTELIGENCIA ARTIFICIAL
    elif opcion == "3":
        print("\n--- SIMULADOR DE IA (Análisis de emociones) ---")
        if lospsicologos == []:
            print("❌ Primero debes registrar al menos un psicólogo en la opción 1.")
        else:
            nombre = input("Introduce el nombre del psicólogo para evaluar su estado con IA: ")

            # Buscamos si existe recorriendo la lista con el diccionario 'psicologo'
            encontrado = False
            for psicologo in lospsicologos:
                if psicologo["nombre"] == nombre:
                    encontrado = True
                    break # Detiene la búsqueda al encontrarlo para conservar sus datos en la variable 'psicologo'

            if encontrado == True:
                print(f"\n🤖 Conexión establecida con el psicólogo: {psicologo['nombre']}")
                frase = input("Escribe una frase sobre cómo te sientes hoy: ")
                frase = frase.lower() # Convertimos a minúsculas

                # La IA analiza la frase usando IF, ELIF, ELSE y operadores lógicos
                if "triste" in frase or "mal" in frase or "agotado" in frase:
                    print("🤖 IA: Detecto desgaste emocional. Recuerda la importancia del autocuidado.")

                elif "enojado" in frase or "molesto" in frase or "estrés" in frase:
                    print("🤖 IA: Alerta de frustración. Se recomienda una pausa activa y respiración.")

                elif "feliz" in frase or "bien" in frase or "motivado" in frase:
                    print("🤖 IA: ¡Excelente estado! Tu energía positiva ayudará mucho hoy.")

                else:
                    print("🤖 IA: Diagnóstico general: Estado estable. ¡Sigue adelante!")

                print("\n[Análisis terminado. Regresando al menú principal...]")

            else:
                print("❌ Ese psicólogo no se encuentra en la lista 'lospsicologos'.")

    # OPCIÓN 4: SALIR
    elif opcion == "4":
        print("\nCerrando el proyecto de simulación de IA. ¡Hasta pronto, CBTIS 128! 👋")
        break # Rompe el while True principal y apaga el programa por completo

    # VALIDACIÓN DEL MENÚ
    else:
        print("❌ Opción inválida. Por favor, elige un número del 1 al 4.")
```
